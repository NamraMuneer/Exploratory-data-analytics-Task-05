Q.7: Have a look at the distribution of reach from different sources

```python
import matplotlib.pyplot as plt

# Calculate the total reach from each source
reach_home = Instagram_data['From Home'].sum()
reach_hashtags = Instagram_data['From Hashtags'].sum()
reach_explore = Instagram_data['From Explore'].sum()
reach_other = Instagram_data['From Other'].sum()

# Labels for each slice of the pie chart
labels = ['Home', 'Hashtags', 'Explore', 'Other']

# Sizes of each slice of the pie chart
sizes = [reach_home, reach_hashtags, reach_explore, reach_other]

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add a title
plt.title('Distribution of Reach from Different Sources')

# Show the plot
plt.show()

```

Q.10: Have a look at the type of hashtags used in the posts using a wordcloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all hashtags into a single string
all_hashtags = ' '.join(Instagram_data['Hashtags'].dropna())

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_hashtags)

# Plot the WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Hashtags')
plt.show()
```

Q.12: Have a look at the distribution of hashtags to see which hashtag is used the most in all the posts


```python
import pandas as pd
import matplotlib.pyplot as plt

all_hashtags = ' '.join(Instagram_data['Hashtags'].dropna())

# Split the combined string into individual hashtags
hashtags_list = all_hashtags.split()

# Create a DataFrame to count the frequency of each hashtag
hashtags_df = pd.DataFrame(hashtags_list, columns=['Hashtag'])
hashtags_counts = hashtags_df['Hashtag'].value_counts()

# Plot the distribution of hashtags
plt.figure(figsize=(10, 6))
hashtags_counts.head(20).plot(kind='bar')
plt.title('Top 20 Most Used Hashtags')
plt.xlabel('Hashtag')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Q.13: Have a look at the distribution of likes and impressions received from the presence of each hashtag


```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data from the CSV file with specified encoding
Instagram_data = pd.read_csv('Instagram data.csv', encoding='latin1')

# Step 2: Preprocess the data and extract hashtags
data_with_hashtags = Instagram_data.dropna(subset=['Hashtags', 'Likes'])
hashtags_split = data_with_hashtags['Hashtags'].str.split()

# Step 3: Calculate the average number of likes for each hashtag
hashtags_likes_data = []

for i, row in data_with_hashtags.iterrows():
    for hashtag in hashtags_split[i]:
        hashtags_likes_data.append({'Hashtag': hashtag, 'Likes': row['Likes']})

hashtags_likes_df = pd.DataFrame(hashtags_likes_data)
average_likes_per_hashtag = hashtags_likes_df.groupby('Hashtag')['Likes'].mean().sort_values(ascending=False)

# Step 4: Visualize the distribution of likes received from each hashtag
plt.figure(figsize=(12, 6))
average_likes_per_hashtag.head(20).plot(kind='bar')
plt.title('Top 20 Hashtags by Average Likes')
plt.xlabel('Hashtag')
plt.ylabel('Average Likes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data from the CSV file with specified encoding
Instagram_data = pd.read_csv('Instagram data.csv', encoding='latin1')

# Step 2: Preprocess the data and extract hashtags
data_with_hashtags = Instagram_data.dropna(subset=['Hashtags', 'Impressions'])
hashtags_split = data_with_hashtags['Hashtags'].str.split()

# Step 3: Calculate the total impressions for each hashtag
hashtags_impressions_data = []

for i, row in data_with_hashtags.iterrows():
    for hashtag in hashtags_split[i]:
        hashtags_impressions_data.append({'Hashtag': hashtag, 'Impressions': row['Impressions']})

hashtags_impressions_df = pd.DataFrame(hashtags_impressions_data)
total_impressions_per_hashtag = hashtags_impressions_df.groupby('Hashtag')['Impressions'].sum().sort_values(ascending=False)

# Step 4: Visualize the distribution of impressions received from each hashtag
plt.figure(figsize=(12, 6))
total_impressions_per_hashtag.head(20).plot(kind='bar')
plt.title('Top 20 Hashtags by Total Impressions')
plt.xlabel('Hashtag')
plt.ylabel('Total Impressions')
plt.xticks(rotation=45, ha='right')
plt.show()
```
